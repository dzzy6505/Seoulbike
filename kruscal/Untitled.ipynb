```python
import csv
from scipy import stats
def mannwhitney (csv_file, question, attribute):
    with open(csv_file, 'r',encoding ='utf-8-sig') as f:
        reader = csv.DictReader(f)
        group_0 = []
        group_1 = []
        group_2 = []
        group_3 = []
        for row in reader:

            if row[question] == '0':
                int_row=float(row[attribute])
                group_0.append(int_row)
            elif row[question] == '1':
                int_row=float(row[attribute])
                group_1.append(int_row)
            elif row[question] == '2':
                int_row=float(row[attribute])
                group_2.append(int_row)

            elif row[question] == '3':
                int_row=float(row[attribute])
                group_3.append(int_row)
    
    
    g0_vs_g1_s = stats.mannwhitneyu(group_0, group_1 ,alternative='two-sided')[0]
    g0_vs_g1_p = stats.mannwhitneyu(group_0, group_1 ,alternative='two-sided')[1]
    
    g0_vs_g2_s = stats.mannwhitneyu(group_0, group_2 ,alternative='two-sided')[0]
    g0_vs_g2_p = stats.mannwhitneyu(group_0, group_2 ,alternative='two-sided')[1]
    
    g0_vs_g3_s = stats.mannwhitneyu(group_0, group_3 ,alternative='two-sided')[0]
    g0_vs_g3_p = stats.mannwhitneyu(group_0, group_3 ,alternative='two-sided')[1]
    
    g1_vs_g2_s = stats.mannwhitneyu(group_1, group_2 ,alternative='two-sided')[0]
    g1_vs_g2_p = stats.mannwhitneyu(group_1, group_2 ,alternative='two-sided')[1]
    
    g1_vs_g3_s = stats.mannwhitneyu(group_1, group_3 ,alternative='two-sided')[0]
    g1_vs_g3_p = stats.mannwhitneyu(group_1, group_3 ,alternative='two-sided')[1]
    
    g2_vs_g3_s = stats.mannwhitneyu(group_2, group_3 ,alternative='two-sided')[0]
    g2_vs_g3_p = stats.mannwhitneyu(group_2, group_3 ,alternative='two-sided')[1]
    return {'집단0 과 집단1 S': g0_vs_g1_s,'집단0 과 집단1 P': g0_vs_g1_p,
            '집단0 과 집단2 S': g0_vs_g2_s,'집단0 과 집단2 P': g0_vs_g2_p,
            '집단0 과 집단3 S': g0_vs_g3_s, '집단0 과 집단3 P': g0_vs_g3_p,
            '집단1 과 집단2 S': g1_vs_g2_s, '집단1 과 집단2 p': g1_vs_g2_p,
            '집단1 과 집단3 S': g1_vs_g3_s, '집단1 과 집단3 p': g1_vs_g3_p,
            '집단2 과 집단3 S': g2_vs_g3_s,'집단2 과 집단3 P': g2_vs_g3_p,
           }


def total (one, two, three, four, five, six):
    total=[]
    total.append(one)
    total.append(two)
    total.append(three)
    total.append(four)
    total.append(five)
    total.append(six)
    return total
```

```python
import csv
from scipy import stats
f = open('kruscal.csv', 'r',encoding ='utf-8-sig')
reader = csv.DictReader(f)
group0=[]
group1=[]
for row in reader:
    if row['Miles_Group'] == '0':
        int_row=int(row['recog_function2'])
        group0.append(int_row)
    elif row['Miles_Group'] == '1':
        int_row=int(row['recog_function2'])
        group1.append(int_row)

stats.mannwhitneyu(group0, group1, alternative='two-sided')

```
