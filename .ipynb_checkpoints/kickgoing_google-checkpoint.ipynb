##### 수정된 konply

```python
sentences_tag = []
for sentence in c_text:
    #post_data = postprocessor.pos(sentence)
    morph = ctwitter.pos(sentence)
    sentences_tag.append(morph)

```

```python

adj_list = []
for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['Adjective']:
            adj_list.append(word)
```

```python
counts = Counter(adj_list)
print(counts.most_common(40))
```

```python
noun_list = []
for sentence11 in sentences_tag:
    for word, tag in sentence11:
        if tag in ['Noun']:
            noun_list.append(word)
counts = Counter(noun_list)
print(counts.most_common(100))
tags = counts.most_common(60)
```

```python
len(counts)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud

font_path = '/usr/share/fonts/truetype/nanum/NanumMyeongjoBold.ttf'

wc = WordCloud(font_path=font_path, background_color='white', width=800, height=600)
cloud = wc.generate_from_frequencies(dict(tags[2:]))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
```

##### 이모티콘만 추출해보기

```python
df = pd.DataFrame.from_dict(counts, orient='index')
df = df.rename(columns={'index':'words', 0:'count'})
df = df.sort_values(by='count', ascending=False)
df = df[:15]

df["cumpercentage"] = df["count"].cumsum()/df["count"].sum()*100

fig, ax = plt.subplots()
ax.bar(df.index, df["count"], color="C0")
ax2 = ax.twinx()
ax2.plot(df.index, df["cumpercentage"], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
plt.show()
```
